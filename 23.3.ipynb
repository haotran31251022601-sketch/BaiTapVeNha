```python
import folium
from folium.plugins import HeatMap
import random
center = [10.762622, 106.682231]
m = folium.Map(location=center, zoom_start=13)
data_points = []
for _ in range(200):
    vido = center[0] + random.uniform(-0.005, 0.005)
    kinhdo = center[1] + random.uniform(-0.005, 0.005)
    data_points.append([vido, kinhdo])
HeatMap(data_points, radius=20, blur=15, max_zoom=1).add_to(m)
m
#Bai 3
```
