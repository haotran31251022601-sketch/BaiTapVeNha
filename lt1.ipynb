```python
import folium
from folium.plugins import HeatMap
import random
center = [10.762622, 106.682231]
m = folium.Map(location=center, zoom_start=13)
data_points = []
for _ in range(200):
    vido = center[0] + random.uniform(-0.005, 0.005)
    kinhdo = center[1] + random.uniform(-0.005, 0.005)
    data_points.append([vido, kinhdo])
HeatMap(data_points, radius=20, blur=15, max_zoom=1).add_to(m)
m
#Bai 3
```

```python
import folium
center = [10.761142133101485, 106.66837266529853]
m = folium.Map(location=center, zoom_start=13)
folium.Marker(
    location=center,
    popup="Kho hàng trung tâm",
    tooltip="Kho hàng",
    icon=folium.Icon(color="red", icon="home")
).add_to(m)
service_radii = [3, 5, 10]
for r in service_radii:
    folium.Circle(
        location=center,
        radius=r * 1000,
        color="blue" if r == 3 else "green" if r == 5 else "red",
        fill=True,
        fill_opacity=0.2
    ).add_to(m)
m
#Bai5
```
